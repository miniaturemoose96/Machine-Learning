```python
# Import all packages we are required
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.datasets import mnist
from keras.utils import np_utils
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.models import Sequential
import numpy as np
```

```python
# B) Import MINST dataset and split it into testing and training sets as we saw
# in the tutorial. Then, reshape each sample into a row vector and scale it by 255.
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# Reshape to a row vector
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)

# Scale 
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
# C) Perform One hot Encoding for Label y.
# So, your label will be a vector of 10 elements for each data Sample
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
# d) Now Define function called 'model_creator'
# This fucntion will define, create and compile your neural network
# model according to your structure and then return the built model as 
# the output. For the ANN neurons/layers, use the same structure as we had
# in the tutorial
input_size = 784
hidden_neurons = 100
out_size = 10

def model_creator():
    model = Sequential()
    # First Layer input: Does no processing, no need to define input layer
    # Second layer: Hidden Layer
    model.add(Dense(hidden_neurons, input_dim = input_size))# Neurons
    model.add(Activation('sigmoid')) # activation
    # Third Layer: output Layer
    model.add(Dense(out_size, input_dim = hidden_neurons)) # Neurons
    model.add(Activation('softmax')) # activation
    # compile
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model
```

```python
# e) Fix the random state for reproducibility
seed = 2
np.random.seed(seed)
```

```python
# f) use keras to wrap your model as an object
my_keras_model = KerasClassifier(build_fn=model_creator, verbose=2)
# g) Now run SKlearn GridSearch to find the best batch_size
# and epochs. Search in this range: batch_size = [30, 50, 100] epochs = [10, 15, 20]
# In your GridSearch, the estimator is the above model, the scoring should be
# 'neg_log_loss', and you have to use 10-fold CV
batch_size = [30, 50, 100]
epochs = [10, 15, 20]
param_grid = dict(epochs=epochs, batch_size=batch_size)
grid = GridSearchCV(estimator=my_keras_model, param_grid=param_grid, cv=10, scoring='neg_log_loss')
grid_result = grid.fit(X_train, y_train)
print(grid_result.best_score_)
```

```python
# Best batch_size and Epochs
print(grid_result.best_params_)
```

```python
# h) Now, test your model with the best batch_size and epochs on the testing set.
grid_result_testing = grid.fit(X_test, y_test)
print(grid_result_testing.best_score_)
```

```python
# Best Batch size and Epochs for testing data
print(grid_result_testing.best_params_)
```
