```python
# Import Pandas library to read and process data tables
import pandas as pd
import numpy as np
# Randomly splitting the original dataset into training set and testing set
from sklearn.model_selection import train_test_split
# accuracy score will be used to measure the accuracy of our predictions 
from sklearn.metrics import accuracy_score
# import Decision Tree, Logistic Regression and Linear Regression classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils import resample
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier

# part a
# read csv file directly from the web and store it in a Pandas DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/Cancer.csv');
print("PART A");
print("DataFrame");
print(df);
# create a feature and label vector from the DataFrame
# create a python list of feature names that would like to pick from the dataset:
feature_cols = ["Clump_Thickness","Uniformity_of_Cell_Size","Uniformity_of_Cell_Shape","Marginal_Adhesion",
                "Single_Epithelial_Cell_Size","Bare_Nuclei","Bland_Chromatin","Normal_Nucleoli","Mitoses"];
x = df[feature_cols];
y = df['Malignant_Cancer'];
```

```python
# part b
# Split our feature and label vectors into 35-65 splits 
# We're taking 35% of our data for our testing set, and the remaining 65% for our training set. 
dfx_train, dfx_test, dfy_train, dfy_test = train_test_split(x, y, test_size=0.35, random_state=3);
bootstrap_size = int(np.floor(0.8*(df.shape[0])));
```

```python
# part c 
print("PART C");
# train/test and report accuracy of decision tree classifier 
my_DecisionTree = DecisionTreeClassifier(random_state=3);
my_DecisionTree.fit(dfx_train, dfy_train);
dfy_predict_dt = my_DecisionTree.predict(dfx_test);
accuracy_dt = accuracy_score(dfy_test, dfy_predict_dt);
print("Accuracy using Decision Tree");
print(accuracy_dt);
```

```python
# part d
# initiallize 2d array, as well as row and col indices 
prediction_results = [[0 for x in range(19)] for y in range(len(dfy_predict_dt))];
row = 0;
col = 0;
for i in range(19):
    dfx_train_i , dfy_train_i = resample(dfx_train, dfy_train, n_samples = bootstrap_size , random_state=i , replace = True);
    Base_DecisionTree = DecisionTreeClassifier(random_state=3);
    Base_DecisionTree.fit(dfx_train_i, dfy_train_i);
    dfy_predict_dt = Base_DecisionTree.predict(dfx_test);
    # for each value in dfy_predict_dt list add to our matrix incrementing the row by one each time
    for val in dfy_predict_dt:
        prediction_results[row][col] = val;
        row += 1;
    # after iterating through entire list reset row back to 0, move col over by one
    row = 0;
    col += 1;    

# initialize list to hold row value after voting
result_After_Voting = [];
# for each row initialize counter variables to 0
# for each value in a row:
# if the value is 0 increment the 0 counter, else it is 1, increment the 1 counter 
# in the end compare the two counters, add value to result_After_Voting that has the greater count value (ie was present the most)
for r in prediction_results:
    countZeros = 0;
    countOnes = 0;
    for value in r:        
        if value == 0:
            countZeros += 1;
        else:
            countOnes += 1;
    if countZeros > countOnes:
        result_After_Voting.append(0);
    else:
        result_After_Voting.append(1);
        
accuracy_dt = accuracy_score(dfy_test, result_After_Voting);

print("PART D")
print("Result List After Voting")
print(list(result_After_Voting));
print("\nAccuracy using Decision Tree After Voting");
print(accuracy_dt);
```

```python
# part e 
# initialize a random forest classifier, train and test using original training and testing values 
my_RandomForest = RandomForestClassifier(n_estimators = 19, bootstrap = True, random_state=3)
my_RandomForest.fit(dfx_train, dfy_train);
dfy_predict_rf = my_RandomForest.predict(dfx_test);
accuracy_rf = accuracy_score(dfy_test, dfy_predict_rf);

print("PART E")
print("Accuracy Using Random Forest");
print(accuracy_rf);
```
